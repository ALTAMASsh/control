```python
#20. Write a program that determines the roots of a cubic equation using the Cardano formula.
import math

def cardano(a, b, c, d):
    p = (3*a*c - b**2)/(3*a**2)
    q = (2*b**3 - 9*a*b*c + 27*a**2*d)/(27*a**3)
    delta = (q/2)**2 + (p/3)**3
    u = (-q/2 + math.sqrt(delta))**(1/3) if delta >= 0 else (-q/2 + complex(0, math.sqrt(-delta)))**(1/3)
    v = (-q/2 - math.sqrt(delta))**(1/3) if delta >= 0 else (-q/2 - complex(0, math.sqrt(-delta)))**(1/3)
    x1 = u + v - b/(3*a)
    x2 = -(u + v)/2 - b/(3*a) + complex(0, math.sqrt(3)/2)*(u - v)
    x3 = -(u + v)/2 - b/(3*a) - complex(0, math.sqrt(3)/2)*(u - v)
    return x1, x2, x3

a = float(input("Enter the coefficient of x^3: "))
b = float(input("Enter the coefficient of x^2: "))
c = float(input("Enter the coefficient of x: "))
d = float(input("Enter the constant term: "))

x1, x2, x3 = cardano(a, b, c, d)

print(f"The roots of the cubic equation {a}x^3 + {b}x^2 + {c}x + {d} = 0 are:")
print(f"x1 = {x1:.3f}")
print(f"x2 = {x2:.3f}")
print(f"x3 = {x3:.3f}")
```

```python
#24. Create a program that implements a simple text-based adventure game with branching scenarios.
import random

class Game:
    def __init__(self):
        self.player = Player()
        self.current_room = Room("Start Room")

    def play(self):
        while True:
            self.current_room.print_description()
            self.player.print_options()
            choice = self.player.get_input()

            if choice == "exit":
                break

            next_room = self.current_room.get_next_room(choice)
            if next_room is None:
                print("Invalid choice.")
                continue

            self.current_room = next_room

class Player:
    def __init__(self):
        self.inventory = []

    def print_options(self):
        print("What do you want to do?")
        for option in self.current_room.get_options():
            print(f"({option}) {option}")

    def get_input(self):
        return input()

class Room:
    def __init__(self, name):
        self.name = name
        self.description = ""
        self.options = []
        self.next_rooms = {}

    def print_description(self):
        print(self.description)

    def get_options(self):
        return self.options

    def get_next_room(self, choice):
        return self.next_rooms.get(choice)

if __name__ == "__main__":
    game = Game()
    game.play()
```

```python
#25. Build a program that solves a linear equation  for x, considering different cases.
def solve_linear_equation(a, b):
    if a == 0 and b == 0:
        return "Infinite solutions"
    elif a == 0 and b != 0:
        return "No solution"
    else:
        return -b / a
    
print(solve_linear_equation(0, 0)) # Infinite solutions
print(solve_linear_equation(0, 5)) # No solution
print(solve_linear_equation(2, 3)) # -1.5
```

```python
### 32. Develop a program that performs matrix addition and subtraction based on user input
def add_matrices(m1, m2):
   
    if len(m1) != len(m2) or len(m1[0]) != len(m2[0]):
        print("Invalid matrices for addition")
        return None
    
    result = []
   
    for i in range(len(m1)):

        row = []
        for j in range(len(m1[0])):
            
            row.append(m1[i][j] + m2[i][j])
       
        result.append(row)
   
    return result


def subtract_matrices(m1, m2):
   
    if len(m1) != len(m2) or len(m1[0]) != len(m2[0]):
        print("Invalid matrices for subtraction")
        return None
    
    result = []
   
    for i in range(len(m1)):
      
        row = []
        for j in range(len(m1[0])):
            
            row.append(m1[i][j] - m2[i][j])
        
        result.append(row)
    
    return result


rows1 = int(input("Enter the number of rows of the first matrix: "))
cols1 = int(input("Enter the number of columns of the first matrix: "))


matrix1 = []


print("Enter the elements of the first matrix row by row:")
for i in range(rows1):
    
    row = list(map(int, input().split()))
   
    if len(row) != cols1:
        print("Invalid input")
        exit()
    
    matrix1.append(row)


rows2 = int(input("Enter the number of rows of the second matrix: "))
cols2 = int(input("Enter the number of columns of the second matrix: "))


matrix2 = []


print("Enter the elements of the second matrix row by row:")
for i in range(rows2):
  
    row = list(map(int, input().split()))
    
    if len(row) != cols2:
        print("Invalid input")
        exit()
  
    matrix2.append(row)


operation = input("Enter + for addition or - for subtraction: ")


if operation not in ["+", "-"]:
    print("Invalid operation")
    exit()


if operation == "+":
    result = add_matrices(matrix1, matrix2)
    if result is not None:
        print("The sum of the matrices is:")
        for row in result:
            print(*row)
else:
    result = subtract_matrices(matrix1, matrix2)
    if result is not None:
        print("The difference of the matrices is:")
        for row in result:
            print(*row)
```

```python
#34. Build a program that performs matrix multiplication using nested loops and conditional statements
def matrix_multiplication(A, B):
    if len(A[0]) != len(B):
        raise ValueError("Matrices are not multiplication compatible.")
    C = [[0 for i in range(len(B[0]))] for j in range(len(A))]
    for i in range(len(A)):
        for j in range(len(B[0])):
            for k in range(len(B)):
                C[i][j] += A[i][k] * B[k][j]
    return C



A = [[1, 2], [3, 4]]
B = [[5, 6], [7, 8]]
C = matrix_multiplication(A, B)
print(C)  
```
